```python
def plot_wide(f):
    def wrapped():
        rcParams['figure.figsize'] = 20, 10
    return wrapped

def rmse(y_true, y_pred):
    
    return np.sqrt(mean_squared_error(y_true, y_pred))

def month_parse(df, col_conv):
    df['YEAR_ID'] = df[col_conv].astype(str).str.slice(0, 4)
    df['MONTH_ID'] = df[col_conv].astype(str).str.slice(5, 7)
    df[col_conv + '_date'] = pd.to_datetime(df[col_conv])
    df[col_conv] = df['YEAR_ID'].astype(int)*100 + df['MONTH_ID'].astype(int)
    return df.drop(labels=['YEAR_ID', 'MONTH_ID'], axis=1)

def data_convert(df, col_exc, do_print=True):
    
    columns = [column for column in df.columns if column not in [col_exc]]
    
    for column in columns:

        if (df[column].dtype not in ['datetime64[ns]', 'object', 'float']):
            
            mn = df[column].min()
            mx = df[column].max()
            
            #uint
            if (mn >= 0):
                
                #uint8
                if (mx < 255):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint8')
                        if do_print:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint8'
                            )
                    except:
                        print(sys.exc_info()[0])
                
                #uint16
                elif (mx < 65535):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint16')
                        if do_print:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint16'
                            )
                    except:
                        print(sys.exc_info()[0])
                        
                #uint32
                elif (mx < 4294967295):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint32')
                        if do_print:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint32'
                            )
                    except:
                        print(sys.exc_info()[0])
                        
                #uint64
                elif (mx < 18446744073709551615):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint64')
                        if do_print:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint64'
                            )
                    except:
                        print(sys.exc_info()[0])
            
            #int8
            elif ((-128 <= mn <= 127) & (-128 <= mx <= 127)):
                try:
                    column_type = str(df[column].dtype)
                    df[column] = df[column].astype('int8')
                    if do_print:
                        print(
                            column,
                            'with type',
                            column_type,
                            'converted to int8'
                        )
                except:
                    print(sys.exc_info()[0])
                
            #int16
            elif ((-32768 <= mn <= 32767) & (-32768 <= mx <= 32767)):
                try:
                    column_type = str(df[column].dtype)
                    df[column] = df[column].astype('int16')
                    if do_print:
                        print(
                            column,
                            'with type',
                            column_type,
                            'converted to int16'
                        )
                except:
                    print(sys.exc_info()[0])
                
            #int32
            elif ((-2147483648 <= mn <= 2147483647) & (-2147483648 <= mx <= 2147483647)):
                try:
                    column_type = str(df[column].dtype)
                    df[column] = df[column].astype('int32')
                    if do_print:
                        print(
                            column,
                            'with type',
                            column_type,
                            'converted to int32'
                        )
                except:
                    print(sys.exc_info()[0])
        
        elif (df[column].dtype == 'float'):
            
            try:
                column_type = str(df[column].dtype)
                df[column] = df[column].astype('float32')
                if do_print:
                    print(
                        column,
                        'with type',
                        column_type,
                        'converted to float32'
                    )
            except:
                print(sys.exc_info()[0]) 
        
        else:
            if do_print:
                print(
                    column,
                    'has no convertable type'
                )
                
    return df

def model_mix(y0, y1, y2, y3, y4, y5, y_true):
    
    rmse_best = 1000000
    v = {
        'v0': .16666666666666668,
        'v1': .16666666666666668,
        'v2': .16666666666666668,
        'v3': .16666666666666668,
        'v4': .16666666666666668,
        'v5': .16666666666666668
    }
    
    variants = []
    for element in itertools.product(
        range(0, 101),
        range(0, 101),
        range(0, 101),
        range(0, 101),
        range(0, 101),
        range(0, 101)
    ):
        s = 0
        for e in element:
            s += e
        if s == 100:
            variants.append(element)
    
    for var in variants:
        
        y_pred = var[0]/100*y0 \
                    + var[1]/100*y1 \
                    + var[1]/100*y2 \
                    + var[2]/100*y3 \
                    + var[3]/100*y4 \
                    + var[4]/100*y5 \
        
        rmse_ = rmse(y_true, y_pred)
        
        #print(
        #    'RMSE:',
        #    rmse_,
        #    '\nPROPORTION:',
        #    var[0],
        #    '//',
        #    var[1],
        #    '//',
        #    var[2],
        #    '//',
        #    var[3]
        #)
        
        if rmse_best > rmse_:
            
            rmse_best = rmse_
            v['v0'] = var[0]
            v['v1'] = var[1]
            v['v2'] = var[2]
            v['v3'] = var[3]
            v['v4'] = var[3]
            v['v5'] = var[3]
            
    print(
        'RMSE:',
        rmse_best,
        '\nTHE PROPORTIONS:',
        v
    )

    return v
            
def data_desc(df):
    
    df_types = pd.DataFrame(df.dtypes.reset_index()).rename(columns={'index': 'column', 0: 'type'})
    
    for c in df_types['column'].values:
        
        null = 'EXIST' if df[c].isnull().any()==True else 'DOES NOT EXIST'
        column_type = df_types[df_types['column']==c]['type'].values[0]
        
        print(
            'NULL VALUES',
            null,
            'WITH FORMAT',
            column_type,
            'IN COLUMN',
            c
        )
        
    print(df.dtypes)
    return df.describe().T

#@plot_wide
def simple_analysis(df, cat_feat=None, target=''):
    
    features = df.columns
    dt = df.dtypes
    
    sns.heatmap(df.corr())
    plt.show()
    
    columns = pd.DataFrame(df.dtypes).reset_index().rename(columns={'index': 'column', 0: 'type'})
    dig_feat = columns[columns['type']!='object']['column'].values
    
    for d in dig_feat:
        
        sns.distplot(df[d])
        plt.show()
        
    if cat_feat != None:
        
        for c in cat_feat:
            
            #sns.stripplot(
            #    x=target,
            #    y=c,
            #    data=df,
            #    jitter=True,
            #    linewidth=1
            #)
            sns.boxplot(
                x=c,
                y=target,
                data=df
            )
            plt.show()
            
def plot_feat_imp(feat, imp, n_feat):
    
    feat = feat[:n_feat]
    df_temp = pd.DataFrame(feat, columns=['feature'])
    df_temp['imp'] = imp[:n_feat]
    df_temp.sort_values(
        by=['imp'],
        ascending=True,
        inplace=True
    )
    
    n_feat = len(feat)
    
    plt.barh(
        range(n_feat),
        df_temp['imp'],
        align='center'
    )
    
    plt.yticks(np.arange(n_feat), df_temp['feature'])
    plt.xlabel(xlabel('Importance'))
    plt.ylabel('Feature')
    del df_temp; gc.collect()
    
def plot_vc(x, y_train, y_valid):
    '''
    plt.title('Validation curve')
    plt.xlabel('depth')
    plt.ylabel('rmse')
    sns.lineplot(
        x=x,
        y=y_train,
        label='Train'
    )

    sns.lineplot(
        x=x,
        y=y_valid,
        label='Valid'
    )
    
    plt.show()
    '''
    train = Scatter(
        x=x,
        y=y_train,
        name='Train',
        line=dict(
            color='rgba(240, 65, 65, .9)',
            shape='hvh'
        ),
        mode='lines+markers',
        connectgaps=True
    )
    
    valid = Scatter(
        x=x,
        y=y_valid,
        name='Valid',
        line=dict(
            color='rgba(65, 100, 240, .9)',
            shape='hvh'
        ),
        mode='lines+markers',
        connectgaps=True
    )
    
    data = [train, valid]
    
    layout = dict(
        title='Validation curve',
        legend=dict(
            x=.9,
            y=1
        ),
        yaxis = dict(
            title='RMSE',
            autorange=False,
            range=[0, max(y_valid)*1.1],
            #tickformat='0.0%'
        ),
        xaxis = dict(
            title='DEPTH',
            dtick=1
        ),
        showlegend=True
    )
    
    fig = dict(
        data=data,
        layout=layout
    )
    
    iplot(fig)
            
def model_mean(**kwargs):
    
    X_train = kwargs['X_train']
    X_test = kwargs['X_test']
    y_train = kwargs['y_train']
    y_test = kwargs['y_test']
    feat = kwargs['feat']

    if 'cat_feat' in kwargs.keys():
        cat_feat = kwargs['cat_feat']

    model = kwargs['model']
    random_state = kwargs['random_state']
    samples = kwargs['samples']
    info = kwargs['info']
    y_pred = []
    
    for i in range(0, samples):
        if 'cat_feat' in kwargs.keys():
            
            y_pred_ = model(
                X_train,
                X_test,
                y_train,
                y_test,
                feat,
                cat_feat,
                random_state+i
            )
        else:
            y_pred_ = model(
                X_train,
                X_test,
                y_train,
                y_test,
                feat,
                random_state+1
            )
            
        y_pred.append(y_pred_)
    
    y_pred = np.mean(y_pred, axis=0)
    
    print(
        samples,
        info,
        'MODELS',
        '\nRMSE:',
        rmse(y_test, y_pred)
    )
   
    return y_pred
```

```python
def merch_prepare(df):
    
    df['most_recent_sales_range'].replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5], inplace=True)
    df['most_recent_purchases_range'].replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5], inplace=True)
    df['avg_sales_lag3'].fillna(0, inplace=True)
    df['avg_sales_lag6'].fillna(0, inplace=True)
    df['avg_sales_lag12'].fillna(0, inplace=True)
    
    df = df[(df['avg_sales_lag3']<100) & (df['avg_sales_lag6']<100) & (df['avg_sales_lag12']<10)]
    
    df['category_2'].fillna(0.0, inplace=True)
    df['category_2'] = df['category_2'].astype('int8')
    df['category_1'].replace(['N', 'Y'], [0, 1], inplace=True)
    df['category_4'].replace(['N', 'Y'], [0, 1], inplace=True)
    #df['avg_purchases_lag3'].replace([np.inf], 1e6, inplace=True)
    #df['avg_purchases_lag6'].replace([np.inf], 1e6, inplace=True)
    #df['avg_purchases_lag12'].replace([np.inf], 1e6, inplace=True)
    
    df = df[(df['avg_purchases_lag3']<3) & (df['avg_purchases_lag6']<3) & (df['avg_purchases_lag12']<3)]
    
    #df['avg_sales_lag3'] = np.log1p(df['avg_sales_lag3']+abs(min(df['avg_sales_lag3'])))
    #df['avg_sales_lag6'] = np.log1p(df['avg_sales_lag6']+abs(min(df['avg_sales_lag6'])))
    #df['avg_sales_lag12'] = np.log1p(df['avg_sales_lag12']+abs(min(df['avg_sales_lag12'])))
    #df['avg_purchases_lag3'] = np.power(df['avg_purchases_lag3'], -1)
    #df['avg_purchases_lag6'] = np.power(df['avg_purchases_lag6'], -1)
    #df['avg_purchases_lag12'] = np.power(df['avg_purchases_lag12'], -1)
    #df['numerical_1'] = np.power(df['numerical_1']+abs(min(df['numerical_1']))+1, -1)
    #df['numerical_2'] = np.power(df['numerical_2']+abs(min(df['numerical_2']))+1, -1)
    
    feat = list(df.columns)
    cat_feat = ['category_1', 'category_2', 'category_4', 'most_recent_sales_range', 'most_recent_purchases_range']

    for c in cat_feat:

        dummy = pd.get_dummies(df[c], prefix=c)
        df = pd.concat([df, dummy], axis=1)
        df.drop(labels=[c], axis=1, inplace=True)
    
    return df
```

```python
def feat_eng_hor(df):
    
    df['hist_auth_first_buy'] = (df['hist_auth_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['hist_no_auth_first_buy'] = (df['hist_no_auth_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['hist_auth_last_buy'] = (df['hist_auth_purchase_date_max'] - df['first_active_month_date']).dt.days
    df['hist_no_auth_last_buy'] = (df['hist_no_auth_purchase_date_max'] - df['first_active_month_date']).dt.days
    df['new_first_buy'] = (df['new_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['new_last_buy'] = (df['new_purchase_date_max'] - df['first_active_month_date']).dt.days

    date_features=[
        'hist_auth_purchase_date_max',
        'hist_auth_purchase_date_min',
        'hist_no_auth_purchase_date_max',
        'hist_no_auth_purchase_date_min',
        'new_purchase_date_max',
        'new_purchase_date_min'
    ]

    for f in date_features:
        df[f] = df[f].astype(np.int64) * 1e-9

    df['card_id_total'] = df['new_card_id_size'] + df['hist_auth_card_id_size'] + df['hist_no_auth_card_id_size']
    df['card_id_cnt_total'] = df['new_card_id_count'] + df['hist_auth_card_id_count'] + df['hist_no_auth_card_id_count']
    df['card_id_cnt_ratio'] = df['new_card_id_count'] / (df['hist_auth_card_id_count'] + df['hist_no_auth_card_id_count'])
    df['purchase_amount_total'] = df['new_purchase_amount_sum'] + df['hist_auth_purchase_amount_sum'] + df['hist_no_auth_purchase_amount_sum']
    df['purchase_amount_mean'] = df['new_purchase_amount_mean'] + df['hist_auth_purchase_amount_mean'] + df['hist_no_auth_purchase_amount_mean']
    df['purchase_amount_max'] = df['new_purchase_amount_max'] + df['hist_auth_purchase_amount_max'] + df['hist_no_auth_purchase_amount_max']
    df['purchase_amount_min'] = df['new_purchase_amount_min'] + df['hist_auth_purchase_amount_min'] + df['hist_no_auth_purchase_amount_min']
    df['purchase_amount_ratio'] = df['new_purchase_amount_sum']/(df['hist_auth_purchase_amount_sum'] + df['hist_no_auth_purchase_amount_sum'])
    df['month_diff_mean'] = df['new_month_diff_mean'] + df['hist_auth_month_diff_mean'] + df['hist_no_auth_month_diff_mean']
    df['month_diff_ratio'] = df['new_month_diff_mean'] / (df['hist_auth_month_diff_mean'] + df['hist_no_auth_month_diff_mean'])
    df['month_lag_mean'] = df['new_month_lag_mean'] + df['hist_auth_month_lag_mean'] + df['hist_no_auth_month_lag_mean']
    df['month_lag_max'] = df['new_month_lag_max'] + df['hist_auth_month_lag_max'] + df['hist_no_auth_month_lag_max']
    df['month_lag_min'] = df['new_month_lag_min'] + df['hist_auth_month_lag_min'] + df['hist_no_auth_month_lag_min']
    #df['category_1_mean'] = df['new_category_1_mean'] + df['hist_auth_category_1_mean'] + df['hist_no_auth_category_1_mean']
    df['installments_total'] = df['new_installments_sum'] + df['hist_auth_installments_sum'] + df['hist_no_auth_installments_sum']
    df['installments_mean'] = df['new_installments_mean'] + df['hist_auth_installments_mean'] + df['hist_no_auth_installments_mean']
    df['installments_max'] = df['new_installments_max'] + df['hist_auth_installments_max'] + df['hist_no_auth_installments_max']
    df['installments_ratio'] = df['new_installments_sum'] / (df['hist_auth_installments_sum'] + df['hist_no_auth_installments_sum'])
    df['price_total'] = df['purchase_amount_total'] / df['installments_total']
    df['price_mean'] = df['purchase_amount_mean'] / df['installments_mean']
    df['price_max'] = df['purchase_amount_max'] / df['installments_max']
    df['duration_mean'] = df['new_duration_mean'] + df['hist_auth_duration_mean'] + df['hist_no_auth_duration_mean']
    df['duration_min'] = df['new_duration_min'] + df['hist_auth_duration_min'] + df['hist_no_auth_duration_min']
    df['duration_max'] = df['new_duration_max'] + df['hist_auth_duration_max'] + df['hist_no_auth_duration_max']
    df['amount_month_ratio_mean'] = df['new_amount_month_ratio_mean'] + df['hist_auth_amount_month_ratio_mean'] + df['hist_no_auth_amount_month_ratio_mean']
    df['amount_month_ratio_min'] = df['new_amount_month_ratio_min'] + df['hist_auth_amount_month_ratio_min'] + df['hist_no_auth_amount_month_ratio_min']
    df['amount_month_ratio_max'] = df['new_amount_month_ratio_max'] + df['hist_auth_amount_month_ratio_max'] + df['hist_no_auth_amount_month_ratio_max']
    df['new_CLV'] = df['new_card_id_count'] * df['new_purchase_amount_sum'] / df['new_month_diff_mean']
    df['hist_auth_CLV'] = df['hist_auth_card_id_count'] * df['hist_auth_purchase_amount_sum'] / (df['hist_auth_month_diff_mean'] + df['hist_no_auth_month_diff_mean'])
    df['CLV_ratio'] = df['new_CLV'] / df['hist_auth_CLV']
    
    return df

def feat_eng_ver(df):
    
    df['full_first_buy'] = (df['full_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['full_last_buy'] = (df['full_purchase_date_max'] - df['first_active_month_date']).dt.days
    df['full_first_buy'] = (df['full_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['full_last_buy'] = (df['full_purchase_date_max'] - df['first_active_month_date']).dt.days

    date_features=[
        'full_purchase_date_max',
        'full_purchase_date_min'
    ]

    for f in date_features:
        df[f] = df[f].astype(np.int64) * 1e-9

    df['full_price_total'] = df['full_purchase_amount_sum'] / df['full_installments_sum']
    df['full_price_mean'] = df['full_purchase_amount_mean'] / df['full_installments_mean']
    df['full_price_max'] = df['full_purchase_amount_max'] / df['full_installments_max']
    df['full_CLV'] = df['full_card_id_count'] * df['full_purchase_amount_sum'] / df['full_month_diff_mean']
    
    return df
```

## 13.8 Ensembling

```python
%%time

train_stack = []
test_stack = []

train_stack.append(X_train_full['y_pred_rfr_mix'])
train_stack.append(X_train_full['y_pred_etr_mix'])
train_stack.append(X_train_full['y_pred_lgbm_r_trees_mix'])
train_stack.append(X_train_full['y_pred_lgbm_r_goss_mix'])
train_stack.append(X_train_full['y_pred_lgbm_r_dart_mix'])
train_stack.append(X_train_full['y_pred_ctbst_r_mix'])

test_stack.append(X_test_full['y_pred_rfr_mix'])
test_stack.append(X_test_full['y_pred_etr_mix'])
test_stack.append(X_test_full['y_pred_lgbm_r_trees_mix'])
test_stack.append(X_test_full['y_pred_lgbm_r_goss_mix'])
test_stack.append(X_test_full['y_pred_lgbm_r_dart_mix'])
test_stack.append(X_test_full['y_pred_ctbst_r_mix'])

train_stack = np.vstack(train_stack).T
test_stack = np.vstack(test_stack).T

folds = KFold(
    n_splits=20,
    shuffle=True,
    random_state=random_state
)
oof_stack = np.zeros(train_stack.shape[0])
pred_stack = np.zeros(test_stack.shape[0])

#for fold_, (trn_idx, val_idx) in enumerate(folds.split(train_stack, y_train)):
for (trn_idx, val_idx) in tqdm_notebook(folds.split(train_stack, y_train)):
    
    #print('fold n{}'.format(fold_))
    trn_data, trn_y = train_stack[trn_idx], y_train.iloc[trn_idx].values
    val_data, val_y = train_stack[val_idx], y_train.iloc[val_idx].values
    
    #print('-'*10 + 'Stacking' + str(fold_) + '-'*10)
    clf = BayesianRidge()
    clf.fit(
        X=trn_data,
        y=trn_y
    )
    
    oof_stack[val_idx] = clf.predict(val_data)
    #print('CV score fold:', rmse(y_train[val_idx], oof_stack[val_idx]))
    
    pred_stack += clf.predict(test_stack)/folds.n_splits
    
print('CV score summary:', rmse(y_train, oof_stack))
```

```python
def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)) 

initialaizer = keras.initializers.he_normal(seed=random_state)
activation = 'tanh'
optimizer = optimizers.Adam(lr=.1)
loss = 'mse'
counts = 128 #len(feat)
epochs = 100

model = Sequential()
#1
model.add(
    LSTM(
        counts,
        init=initialaizer,
        input_shape=(
            X_train_n.shape[1],
            X_train_n.shape[2]
        )
    )
)
#2
model.add(BatchNormalization(axis=1))
model.add(Dropout(.05))
model.add(Dense(counts//2))
#3
model.add(PReLU())
model.add(BatchNormalization(axis=1))
model.add(Dropout(.03))
model.add(Dense(counts//4))
#4
model.add(PReLU())
model.add(BatchNormalization(axis=1))
model.add(Dropout(.01))
model.add(Dense(counts//8))

model.compile(
    optimizer=optimizer,
    loss='mse',
    metric=[root_mean_squared_error]
)

model.fit(
    x=X_train_n,
    y=y_train_n,
    nb_epoch=epochs,
    batch_size=512,
    verbose=1,
    validation_split=.05,
    #validation_data=(X_valid_n_v, y_valid_n),
    shuffle=True,
    callbacks=None,
    sample_weight=None
)
```
