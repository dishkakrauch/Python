# Set functions

```python
def plot_wide(f):
    def wrapped():
        rcParams['figure.figsize'] = 20, 10
    return wrapped

def rmse(y_true, y_pred):
    
    return np.sqrt(mean_squared_error(y_true, y_pred))

def month_parse(df, col_conv):
    df['YEAR_ID'] = df[col_conv].astype(str).str.slice(0, 4)
    df['MONTH_ID'] = df[col_conv].astype(str).str.slice(5, 7)
    df[col_conv + '_date'] = pd.to_datetime(df[col_conv])
    df[col_conv] = df['YEAR_ID'].astype(int)*100 + df['MONTH_ID'].astype(int)
    return df.drop(labels=['YEAR_ID', 'MONTH_ID'], axis=1)

def data_convert(df, col_exc, do_print=True):
    
    columns = [column for column in df.columns if column not in [col_exc]]
    
    for column in columns:

        if (df[column].dtype not in ['datetime64[ns]', 'object', 'float']):
            
            mn = df[column].min()
            mx = df[column].max()
            
            #uint
            if (mn>=0):
                
                #uint8
                if (mx<255):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint8')
                        if do_print == True:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint8'
                            )
                    except:
                        print(sys.exc_info()[0])
                
                #uint16
                elif (mx<65535):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint16')
                        if do_print == True:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint16'
                            )
                    except:
                        print(sys.exc_info()[0])
                        
                #uint32
                elif (mx<4294967295):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint32')
                        if do_print == True:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint32'
                            )
                    except:
                        print(sys.exc_info()[0])
                        
                #uint64
                elif (mx<18446744073709551615):
                    try:
                        column_type = str(df[column].dtype)
                        df[column] = df[column].astype('uint64')
                        if do_print == True:
                            print(
                                column,
                                'with type',
                                column_type,
                                'converted to uint64'
                            )
                    except:
                        print(sys.exc_info()[0])
            
            #int8
            elif ((-128 <= mn <= 127) & (-128 <= mx <= 127)):
                try:
                    column_type = str(df[column].dtype)
                    df[column] = df[column].astype('int8')
                    if do_print == True:
                        print(
                            column,
                            'with type',
                            column_type,
                            'converted to int8'
                        )
                except:
                    print(sys.exc_info()[0])
                
            #int16
            elif ((-32768 <= mn <= 32767) & (-32768 <= mx <= 32767)):
                try:
                    column_type = str(df[column].dtype)
                    df[column] = df[column].astype('int16')
                    if do_print == True:
                        print(
                            column,
                            'with type',
                            column_type,
                            'converted to int16'
                        )
                except:
                    print(sys.exc_info()[0])
                
            #int32
            elif ((-2147483648 <= mn <= 2147483647) & (-2147483648 <= mx <= 2147483647)):
                try:
                    column_type = str(df[column].dtype)
                    df[column] = df[column].astype('int32')
                    if do_print == True:
                        print(
                            column,
                            'with type',
                            column_type,
                            'converted to int32'
                        )
                except:
                    print(sys.exc_info()[0])
        
        elif (df[column].dtype=='float'):
            
            try:
                column_type = str(df[column].dtype)
                df[column] = df[column].astype('float32')
                if do_print == True:
                    print(
                        column,
                        'with type',
                        column_type,
                        'converted to float32'
                    )
            except:
                print(sys.exc_info()[0]) 
        
        else:
            if do_print == True:
                print(
                    column,
                    'has no convertable type'
                )
                
    return df

def model_mix(y1, y2, y3, y4, y_true):
    
    rmse_best = 1000000
    v = {
        'v0': .25,
        'v1': .25,
        'v2': .25,
        'v3': .25
    }
    
    variants = []
    for element in itertools.product(range(0, 101), range(0, 101), range(0, 101), range(0, 101)):
        s = 0
        for e in element:
            s += e
        if s == 100:
            variants.append(element)
    
    for var in variants:
        
        y_pred = var[0]/100*y1 + var[1]/100*y2 + var[2]/100*y3 + var[3]/100*y4
        rmse_ = rmse(y_true, y_pred)
        
        #print(
        #    'RMSE:',
        #    rmse_,
        #    '\nPROPORTION:',
        #    var[0],
        #    '//',
        #    var[1],
        #    '//',
        #    var[2],
        #    '//',
        #    var[3]
        #)
        
        if rmse_best > rmse_:
            
            rmse_best = rmse_
            v['v0'] = var[0]
            v['v1'] = var[1]
            v['v2'] = var[2]
            v['v3'] = var[3]
    print(
        'RMSE:',
        rmse_best,
        '\nTHE PROPORTIONS:',
        v
    )
            
    return v
            
def data_desc(df):
    
    df_types = pd.DataFrame(df.dtypes.reset_index()).rename(columns={'index': 'column', 0: 'type'})
    
    for c in df_types['column'].values:
        
        null = 'EXIST' if df[c].isnull().any()==True else 'DOES NOT EXIST'
        column_type = df_types[df_types['column']==c]['type'].values[0]
        
        print(
            'NULL VALUES',
            null,
            'WITH FORMAT',
            column_type,
            'IN COLUMN',
            c
        )
        
    print(df.dtypes)
    return df.describe().T

#@plot_wide
def simple_analysis(df, cat_feat=None, target=''):
    
    features = df.columns
    dt = df.dtypes
    
    sns.heatmap(df.corr())
    plt.show()
    
    columns = pd.DataFrame(df.dtypes).reset_index().rename(columns={'index': 'column', 0: 'type'})
    dig_feat = columns[columns['type']!='object']['column'].values
    
    for d in dig_feat:
        
        sns.distplot(df[d])
        plt.show()
        
    if cat_feat != None:
        
        for c in cat_feat:
            
            #sns.stripplot(
            #    x=target,
            #    y=c,
            #    data=df,
            #    jitter=True,
            #    linewidth=1
            #)
            sns.boxplot(
                x=c,
                y=target,
                data=df
            )
            plt.show()
            
def plot_feat_imp(feat, imp):
    
    df_temp = pd.DataFrame(feat, columns=['feature'])
    df_temp['imp'] = imp
    df_temp.sort_values(
        by=['imp'],
        ascending=True,
        inplace=True
    )
    
    n_feat = len(feat)
    
    plt.barh(
        range(n_feat),
        df_temp['imp'],
        align='center'
    )
    
    plt.yticks(np.arange(n_feat), df_temp['feature'])
    plt.xlabel(xlabel('Importance'))
    plt.ylabel('Feature')
    del df_temp
            
def model_mean(**kwargs):
    
    X_train = kwargs['X_train']
    X_test = kwargs['X_test']
    y_train = kwargs['y_train']
    y_test = kwargs['y_test']
    feat = kwargs['feat']

    if 'cat_feat' in kwargs.keys():
        cat_feat = kwargs['cat_feat']

    model = kwargs['model']
    random_state = kwargs['random_state']
    samples = kwargs['samples']
    info = kwargs['info']
    y_pred = []
    
    for i in range(0, samples):
        if 'cat_feat' in kwargs.keys():
            
            y_pred_ = model(
                X_train,
                X_test,
                y_train,
                y_test,
                feat,
                cat_feat,
                random_state+i
            )
        else:
            
            y_pred_ = model(
                X_train,
                X_test,
                y_train,
                y_test,
                feat,
                random_state+1
            )
            
        y_pred.append(y_pred_)
    
    y_pred = np.mean(y_pred, axis=0)
    
    print(
        samples,
        info,
        'MODELS',
        '\nRMSE:',
        rmse(y_test, y_pred)
    )
   
    return y_pred
```

```python
def data_prepare_(df):
    
    df['most_recent_sales_range'].replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5], inplace=True)
    df['most_recent_purchases_range'].replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5], inplace=True)
    df['avg_sales_lag3'].fillna(0, inplace=True)
    df['avg_sales_lag6'].fillna(0, inplace=True)
    df['avg_sales_lag12'].fillna(0, inplace=True)
    
    df = df[(df['avg_sales_lag3']<100) & (df['avg_sales_lag6']<100) & (df['avg_sales_lag12']<10)]
    
    df['category_2'].fillna(0.0, inplace=True)
    df['category_2'] = df['category_2'].astype('int8')
    df['category_1'].replace(['N', 'Y'], [0, 1], inplace=True)
    df['category_4'].replace(['N', 'Y'], [0, 1], inplace=True)
    #df['avg_purchases_lag3'].replace([np.inf], 1e6, inplace=True)
    #df['avg_purchases_lag6'].replace([np.inf], 1e6, inplace=True)
    #df['avg_purchases_lag12'].replace([np.inf], 1e6, inplace=True)
    
    df = df[(df['avg_purchases_lag3']<3) & (df['avg_purchases_lag6']<3) & (df['avg_purchases_lag12']<3)]
    
    #df['avg_sales_lag3'] = np.log1p(df['avg_sales_lag3']+abs(min(df['avg_sales_lag3'])))
    #df['avg_sales_lag6'] = np.log1p(df['avg_sales_lag6']+abs(min(df['avg_sales_lag6'])))
    #df['avg_sales_lag12'] = np.log1p(df['avg_sales_lag12']+abs(min(df['avg_sales_lag12'])))
    #df['avg_purchases_lag3'] = np.power(df['avg_purchases_lag3'], -1)
    #df['avg_purchases_lag6'] = np.power(df['avg_purchases_lag6'], -1)
    #df['avg_purchases_lag12'] = np.power(df['avg_purchases_lag12'], -1)
    #df['numerical_1'] = np.power(df['numerical_1']+abs(min(df['numerical_1']))+1, -1)
    #df['numerical_2'] = np.power(df['numerical_2']+abs(min(df['numerical_2']))+1, -1)
    
    feat = list(df.columns)
    cat_feat = ['category_1', 'category_2', 'category_4', 'most_recent_sales_range', 'most_recent_purchases_range']

    for c in cat_feat:

        dummy = pd.get_dummies(df[c], prefix=c)
        df = pd.concat([df, dummy], axis=1)
        df.drop(labels=[c], axis=1, inplace=True)
    
    return df
```

```python
def feat_eng(df):
    
    df['hist_auth_first_buy'] = (df['hist_auth_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['hist_no_auth_first_buy'] = (df['hist_no_auth_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['hist_auth_last_buy'] = (df['hist_auth_purchase_date_max'] - df['first_active_month_date']).dt.days
    df['hist_no_auth_last_buy'] = (df['hist_no_auth_purchase_date_max'] - df['first_active_month_date']).dt.days
    df['new_first_buy'] = (df['new_purchase_date_min'] - df['first_active_month_date']).dt.days
    df['new_last_buy'] = (df['new_purchase_date_max'] - df['first_active_month_date']).dt.days

    date_features=[
        'hist_auth_purchase_date_max',
        'hist_auth_purchase_date_min',
        'hist_no_auth_purchase_date_max',
        'hist_no_auth_purchase_date_min',
        'new_purchase_date_max',
        'new_purchase_date_min'
    ]

    for f in date_features:
        df[f] = df[f].astype(np.int64) * 1e-9

    df['card_id_total'] = df['new_card_id_size'] + df['hist_auth_card_id_size'] + df['hist_no_auth_card_id_size']
    df['card_id_cnt_total'] = df['new_card_id_count'] + df['hist_auth_card_id_count'] + df['hist_no_auth_card_id_count']
    df['card_id_cnt_ratio'] = df['new_card_id_count'] / (df['hist_auth_card_id_count'] + df['hist_no_auth_card_id_count'])
    df['purchase_amount_total'] = df['new_purchase_amount_sum'] + df['hist_auth_purchase_amount_sum'] + df['hist_no_auth_purchase_amount_sum']
    df['purchase_amount_mean'] = df['new_purchase_amount_mean'] + df['hist_auth_purchase_amount_mean'] + df['hist_no_auth_purchase_amount_mean']
    df['purchase_amount_max'] = df['new_purchase_amount_max'] + df['hist_auth_purchase_amount_max'] + df['hist_no_auth_purchase_amount_max']
    df['purchase_amount_min'] = df['new_purchase_amount_min'] + df['hist_auth_purchase_amount_min'] + df['hist_no_auth_purchase_amount_min']
    df['purchase_amount_ratio'] = df['new_purchase_amount_sum'] / (df['hist_auth_purchase_amount_sum'] + df['hist_no_auth_purchase_amount_sum'])
    df['month_diff_mean'] = df['new_month_diff_mean'] + df['hist_auth_month_diff_mean'] + df['hist_no_auth_month_diff_mean']
    df['month_diff_ratio'] = df['new_month_diff_mean'] / (df['hist_auth_month_diff_mean'] + df['hist_no_auth_month_diff_mean'])
    df['month_lag_mean'] = df['new_month_lag_mean'] + df['hist_auth_month_lag_mean'] + df['hist_no_auth_month_lag_mean']
    df['month_lag_max'] = df['new_month_lag_max'] + df['hist_auth_month_lag_max'] + df['hist_no_auth_month_lag_max']
    df['month_lag_min'] = df['new_month_lag_min'] + df['hist_auth_month_lag_min'] + df['hist_no_auth_month_lag_min']
    #df['category_1_mean'] = df['new_category_1_mean'] + df['hist_auth_category_1_mean'] + df['hist_no_auth_category_1_mean']
    df['installments_total'] = df['new_installments_sum'] + df['hist_auth_installments_sum'] + df['hist_no_auth_installments_sum']
    df['installments_mean'] = df['new_installments_mean'] + df['hist_auth_installments_mean'] + df['hist_no_auth_installments_mean']
    df['installments_max'] = df['new_installments_max'] + df['hist_auth_installments_max'] + df['hist_no_auth_installments_max']
    df['installments_ratio'] = df['new_installments_sum'] / (df['hist_auth_installments_sum'] + df['hist_no_auth_installments_sum'])
    df['price_total'] = df['purchase_amount_total'] / df['installments_total']
    df['price_mean'] = df['purchase_amount_mean'] / df['installments_mean']
    df['price_max'] = df['purchase_amount_max'] / df['installments_max']
    df['duration_mean'] = df['new_duration_mean'] + df['hist_auth_duration_mean'] + df['hist_no_auth_duration_mean']
    df['duration_min'] = df['new_duration_min'] + df['hist_auth_duration_min'] + df['hist_no_auth_duration_min']
    df['duration_max'] = df['new_duration_max'] + df['hist_auth_duration_max'] + df['hist_no_auth_duration_max']
    df['amount_month_ratio_mean'] = df['new_amount_month_ratio_mean'] + df['hist_auth_amount_month_ratio_mean'] + df['hist_no_auth_amount_month_ratio_mean']
    df['amount_month_ratio_min'] = df['new_amount_month_ratio_min'] + df['hist_auth_amount_month_ratio_min'] + df['hist_no_auth_amount_month_ratio_min']
    df['amount_month_ratio_max'] = df['new_amount_month_ratio_max'] + df['hist_auth_amount_month_ratio_max'] + df['hist_no_auth_amount_month_ratio_max']
    df['new_CLV'] = df['new_card_id_count'] * df['new_purchase_amount_sum'] / df['new_month_diff_mean']
    df['hist_auth_CLV'] = df['hist_auth_card_id_count'] * df['hist_auth_purchase_amount_sum'] / (df['hist_auth_month_diff_mean'] + df['hist_no_auth_month_diff_mean'])
    df['CLV_ratio'] = df['new_CLV'] / df['hist_auth_CLV']
    
    return df
```

```python
def model_LGBM(X_train, X_test, y_train, y_test, feat, cat_feat, random_state):
    
    X_train_lgb, X_valid_lgb, y_train_lgb, y_valid_lgb = train_test_split(
        X_train[feat],
        y_train,
        test_size=.25,
        random_state=random_state
    )

    train_lgb = lgb.Dataset(
        data=X_train_lgb,
        label=y_train_lgb,
        reference=None,
        weight=None,
        group=None,
        init_score=None,
        silent=None,
        feature_name=feat,
        categorical_feature=cat_feat,
        params=None,
        free_raw_data=False
    )

    valid_lgb = lgb.Dataset(
        data=X_valid_lgb,
        label=y_valid_lgb,
        reference=None,
        weight=None,
        group=None,
        init_score=None,
        silent=None,
        feature_name=feat,
        categorical_feature=cat_feat,
        params=None,
        free_raw_data=False
    )

    test_lgb = lgb.Dataset(
        data=X_test[feat],
        label=y_test,
        reference=None,
        weight=None,
        group=None,
        init_score=None,
        silent=None,
        feature_name=feat,
        categorical_feature=cat_feat,
        params=None,
        free_raw_data=False
    )

    global PARAMS_PREV

    def lgbm_callback(is_changing=False):

        def callback(env):

            global PARAMS_PREV

            delta = .0005
            up = 25
            down = 5

            model_callback = env.model
            rmse_callback = rmse(test_lgb.label, model_callback.predict(test_lgb.data))
            
            #print(env.iteration)
            
            #up each nth iteration
            if (PARAMS_PREV['learning_rate'] + delta < .1) & \
                                    (env.iteration % up == 0) & \
                                    (env.iteration > 0) & \
                                    (is_changing == True):

                PARAMS_PREV['learning_rate'] += delta

                print(
                    '\nRESET LEARNING RATE UP TO:',
                      params_prev['learning_rate']
                )

                print(
                   'ITERATION:',
                    env.model.current_iteration(),
                    'RMSE:',
                    rmse_callback
                )

            #down each nth iteration
            elif (PARAMS_PREV['learning_rate'] - delta > .0001) & \
                                    (env.iteration % down == 0) & \
                                    (env.iteration > 0) & \
                                    (is_changing == True):

                PARAMS_PREV['learning_rate'] -= delta

                print(
                    '\nRESET LEARNING RATE DOWN TO:',
                      PARAMS_PREV['learning_rate']
                )

                print(
                   'ITERATION:',
                    env.model.current_iteration(),
                    'RMSE:',
                    rmse_callback
                )

        callback.before_iteration = True
        callback.order = 0
        return callback
    
    params = {
    #Core Parameters
        #'config': '',
        'task': 'train',
        'objective': 'regression',
        'boosting': 'gbrt',
        #'data': '',
        #'valid': '',
        #num_iterations: 100000,
        'learning_rate': .01,
        'num_leaves': 716, #2^max_depth*70%
        #'tree_learner': 'serial',
        'num_threads': n_threads,
        'device_type': 'cpu',
        'seed': random_state,
    #Learning Control Parameters
        'max_depth': 8,
        'min_data_in_leaf': 20,
        'min_sum_hessian_in_leaf': 1e-3,
        'bagging_fraction': .75,
        'bagging_freq': 2,
        'bagging_seed': random_state,
        'feature_fraction': .75,
        'feature_fraction_seed': random_state,
        #'early_stopping_rounds': 100,
        'max_delta_step': 0.0,
        'lambda_l1': 0.0,
        'lambda_l2': 0.0,
        'min_gain_to_split': 0.0,
        #'drop_rate': .1, #dart
        #'max_dop': 50, #dart
        #skip_drop': .5, #dart
        #'xgboost_dart_mode': False, #dart
        #'uniform_drop': False, #dart
        #'drop_seed': 4, #dart
        #'top_rate': .2, #goss
        #'oter_rate': .1, #goss
        'min_data_per_group': 100,
        'max_cat_threshold': 32,
        'cat_l2': 10.0,
        'cat_smooth': 10.0,
        'max_cat_to_onehot': 4,
        #'top_k': 20, #parallel
        'monotone_constraints': None,
        'feature_contri': None,
        'forcedsplits_filename': '',
        'refit_decay_rate': .9,
    #IO Parameters
        'verbosity': 1,
        'max_bin': 133,
        'min_data_in_bin': 3,
        'bin_construct_sample_cnt': 200000,
        'histogram_pool_size': -1.0,
        'data_random_seed': random_state,
        #'output_model': LightGBM_mode.txt, #model
        'snapshot_freq': -1,
        #'input_model': '', #model
        #'output_result': LightGBM_predict_result.txt, #model
        'iniscore_filename': '',
        'valid_data_iniscore': '',
        'pre_partition': False,
        'enable_bundle': True,
        'max_conflict': 0.0,
        'is_enable_sparse': True,
        'sparse_threshold': .8,
        'use_missing': True,
        'zero_as_missing': True,
        'two_round': False,
        'save_binary': False,
        'enable_load_from_binary_file': True,
        'header': False,
        'label_column': '',
        'weight_columns': '',
        'group_column': '',
        'ignore_columns': '',
        #'categorial_feature': '', #train
        #'predict_raw_score': False, #predict
        #'predict_leaf_index': False, #predict
        #'predict_contrlib': False, #predict
        #'num_iterations_predict': -1, #predict
        #'pred_early_stop': False, #predict
        #'pred_early_stop_freq': 10, #predict
        #'pred_early_stop_margin': 10.0, #predict
        #'convert_model_language': '', #convert_model
        #'convert_model': 'gbdt_prediction_cpp', #convert_model
    #Objective Parameters
        #'num_class': 1, #multi_class
        #'is_unbalanced': False, #binary
        #'scale_pos_weight': 1.0, #binary
        #'sigmoid': 1.0, #binary
        'boost_from_average': True,
        'reg_sqrt': False,
        'alpha': .9,
        'fair_c': 1.0,
        'poisson_max_delta_step': .7,
        'tweedie_variance_power': 1.5,
        #'max_position': 20, #lambdarank
        #'label_gain': 0,1,3,7,15,31,63,...,2^30-1, #lambdarank
    #Metric Parameters
        'metric': 'l2_root',
        'metric_freq': 1,
        'is_provide_training_metric': False,
    #Network Parameters
        'num_machines': 1,
        'local_listen_port': 12400,
        'time-out': 120,
        'machine_list_filename': '',
        'machines': '',
    #GPU Parameters
        'gpu_platform_id': 0,
        'gpu_device_id': 0,
        'gpu_use_dp': True
    }

    PARAMS_PREV = {'learning_rate': params['learning_rate']}
    
    model = lgb.train(
        params=params,
        train_set=train_lgb,
        num_boost_round=100000,
        valid_sets=valid_lgb,
        valid_names=None,
        fobj=None,
        feval=None,
        init_model=None,
        feature_name='auto',
        categorical_feature=cat_feat,
        early_stopping_rounds=100,
        evals_result={'train': train_lgb, 'valid': valid_lgb},
        verbose_eval=100,
        learning_rates=None,
        keep_training_booster=False,
        callbacks=[lgbm_callback(False)]
    )

    y_pred = model.predict(
        data=test_lgb.data,
        num_iteration=-1,
        raw_score=False,
        pred_leaf=False,
        pred_contrib=False,
        data_has_header=False,
        is_reshape=True,
        pred_parameter=None
    )

    print(
        'MODEL LGBM',
        '\nRMSE:',
        rmse(test_lgb.label, y_pred)
    )
    
    return y_pred
```

```python
def model_LGBM(X_train, X_test, y_train, y_test, feat, cat_feat, random_state):
    
    X_train_lgb, X_valid_lgb, y_train_lgb, y_valid_lgb = train_test_split(
        X_train[feat],
        y_train,
        test_size=.25,
        random_state=random_state
    )

    train_lgb = lgb.Dataset(
        data=X_train_lgb,
        label=y_train_lgb,
        reference=None,
        weight=None,
        group=None,
        init_score=None,
        silent=None,
        feature_name=feat,
        categorical_feature=cat_feat,
        params=None,
        free_raw_data=False
    )

    valid_lgb = lgb.Dataset(
        data=X_valid_lgb,
        label=y_valid_lgb,
        reference=None,
        weight=None,
        group=None,
        init_score=None,
        silent=None,
        feature_name=feat,
        categorical_feature=cat_feat,
        params=None,
        free_raw_data=False
    )

    test_lgb = lgb.Dataset(
        data=X_test[feat],
        label=y_test,
        reference=None,
        weight=None,
        group=None,
        init_score=None,
        silent=None,
        feature_name=feat,
        categorical_feature=cat_feat,
        params=None,
        free_raw_data=False
    )

    global PARAMS_PREV

    def lgbm_callback(is_changing=False):

        def callback(env):

            global PARAMS_PREV

            delta = .0005
            up = 25
            down = 5

            model_callback = env.model
            rmse_callback = rmse(test_lgb.label, model_callback.predict(test_lgb.data))
            
            #print(env.iteration)
            
            #up each nth iteration
            if (PARAMS_PREV['learning_rate'] + delta < .1) & \
                                    (env.iteration % up == 0) & \
                                    (env.iteration > 0) & \
                                    (is_changing == True):

                PARAMS_PREV['learning_rate'] += delta

                print(
                    '\nRESET LEARNING RATE UP TO:',
                      PARAMS_PREV['learning_rate']
                )

                print(
                   'ITERATION:',
                    env.model.current_iteration(),
                    'RMSE:',
                    rmse_callback
                )

            #down each nth iteration
            elif (PARAMS_PREV['learning_rate'] - delta > .0001) & \
                                    (env.iteration % down == 0) & \
                                    (env.iteration > 0) & \
                                    (is_changing == True):

                PARAMS_PREV['learning_rate'] -= delta

                print(
                    '\nRESET LEARNING RATE DOWN TO:',
                      PARAMS_PREV['learning_rate']
                )

                print(
                   'ITERATION:',
                    env.model.current_iteration(),
                    'RMSE:',
                    rmse_callback
                )

        callback.before_iteration = True
        callback.order = 0
        return callback
    
    params = {
    #Core Parameters
        #'config': '',
        'task': 'train',
        'objective': 'regression',
        'boosting': 'gbrt',
        #'data': '',
        #'valid': '',
        #num_iterations: 100000,
        'learning_rate': .01,
        'num_leaves': 180, #2^max_depth*70%
        #'tree_learner': 'serial',
        'num_threads': n_threads,
        'device_type': 'cpu',
        'seed': random_state,
    #Learning Control Parameters
        'max_depth': 8,
        'min_data_in_leaf': 20,
        'min_sum_hessian_in_leaf': 1e-3,
        'bagging_fraction': .75,
        'bagging_freq': 2,
        'bagging_seed': random_state,
        'feature_fraction': .75,
        'feature_fraction_seed': random_state,
        #'early_stopping_rounds': 100,
        'max_delta_step': 0.0,
        'lambda_l1': 0.0,
        'lambda_l2': 0.0,
        'min_gain_to_split': 0.0,
        #'drop_rate': .1, #dart
        #'max_dop': 50, #dart
        #skip_drop': .5, #dart
        #'xgboost_dart_mode': False, #dart
        #'uniform_drop': False, #dart
        #'drop_seed': 4, #dart
        #'top_rate': .2, #goss
        #'oter_rate': .1, #goss
        'min_data_per_group': 100,
        'max_cat_threshold': 32,
        'cat_l2': 10.0,
        'cat_smooth': 10.0,
        'max_cat_to_onehot': 4,
        #'top_k': 20, #parallel
        'monotone_constraints': None,
        'feature_contri': None,
        'forcedsplits_filename': '',
        'refit_decay_rate': .9,
    #IO Parameters
        'verbosity': 1,
        'max_bin': 133,
        'min_data_in_bin': 3,
        'bin_construct_sample_cnt': 200000,
        'histogram_pool_size': -1.0,
        'data_random_seed': random_state,
        #'output_model': LightGBM_mode.txt, #model
        'snapshot_freq': -1,
        #'input_model': '', #model
        #'output_result': LightGBM_predict_result.txt, #model
        'iniscore_filename': '',
        'valid_data_iniscore': '',
        'pre_partition': False,
        'enable_bundle': True,
        'max_conflict': 0.0,
        'is_enable_sparse': True,
        'sparse_threshold': .8,
        'use_missing': True,
        'zero_as_missing': True,
        'two_round': False,
        'save_binary': False,
        'enable_load_from_binary_file': True,
        'header': False,
        'label_column': '',
        'weight_columns': '',
        'group_column': '',
        'ignore_columns': '',
        #'categorial_feature': '', #train
        #'predict_raw_score': False, #predict
        #'predict_leaf_index': False, #predict
        #'predict_contrlib': False, #predict
        #'num_iterations_predict': -1, #predict
        #'pred_early_stop': False, #predict
        #'pred_early_stop_freq': 10, #predict
        #'pred_early_stop_margin': 10.0, #predict
        #'convert_model_language': '', #convert_model
        #'convert_model': 'gbdt_prediction_cpp', #convert_model
    #Objective Parameters
        #'num_class': 1, #multi_class
        #'is_unbalanced': False, #binary
        #'scale_pos_weight': 1.0, #binary
        #'sigmoid': 1.0, #binary
        'boost_from_average': True,
        'reg_sqrt': False,
        'alpha': .9,
        'fair_c': 1.0,
        'poisson_max_delta_step': .7,
        'tweedie_variance_power': 1.5,
        #'max_position': 20, #lambdarank
        #'label_gain': 0,1,3,7,15,31,63,...,2^30-1, #lambdarank
    #Metric Parameters
        'metric': 'l2_root',
        'metric_freq': 1,
        'is_provide_training_metric': False,
    #Network Parameters
        'num_machines': 1,
        'local_listen_port': 12400,
        'time-out': 120,
        'machine_list_filename': '',
        'machines': '',
    #GPU Parameters
        'gpu_platform_id': 0,
        'gpu_device_id': 0,
        'gpu_use_dp': True
    }

    PARAMS_PREV = {'learning_rate': params['learning_rate']}
    
    model = lgb.train(
        params=params,
        train_set=train_lgb,
        num_boost_round=10000,
        valid_sets=valid_lgb,
        valid_names=None,
        fobj=None,
        feval=None,
        init_model=None,
        feature_name='auto',
        categorical_feature=cat_feat,
        early_stopping_rounds=100,
        evals_result={'train': train_lgb, 'valid': valid_lgb},
        verbose_eval=100,
        learning_rates=None,
        keep_training_booster=False,
        callbacks=[lgbm_callback(False)]
    )

    y_pred = model.predict(
        data=test_lgb.data,
        num_iteration=-1,
        raw_score=False,
        pred_leaf=False,
        pred_contrib=False,
        data_has_header=False,
        is_reshape=True,
        pred_parameter=None
    )

    print(
        'MODEL LGBM',
        '\nRMSE:',
        rmse(test_lgb.label, y_pred)
    )
    
    return y_pred
```
